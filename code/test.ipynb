```python
import sys

max_val = -1
max_row = 0
max_col = 0

for i in range(9):
    row = list(map(int, sys.stdin.readline().split()))
    current_max = max(row)
    if current_max > max_val:
        max_val = current_max
        max_row = i+1
        max_col = row.index(current_max) + 1

print(max_val)
print(max_row, max_col)
```
